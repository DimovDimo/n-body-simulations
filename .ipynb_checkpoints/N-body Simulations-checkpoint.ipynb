```python
import numpy as np
import matplotlib.pyplot as plt

def get_accelerations(pos, mass, G=1.0, softening=0.1):
    """
    Calculate accelerations using Direct Summation O(N^2)
    softening: prevents division by zero and numerical instability
    """
    # Vectorized subtraction: r_ij = pos_j - pos_i
    dx = pos[:, 0:1].T - pos[:, 0:1]
    dy = pos[:, 1:2].T - pos[:, 1:2]
    dz = pos[:, 2:3].T - pos[:, 2:3]
    
    # Distance squared with softening
    inv_r3 = (dx**2 + dy**2 + dz**2 + softening**2)**(-1.5)
    
    ax = G * (dx * inv_r3) @ mass
    ay = G * (dy * inv_r3) @ mass
    az = G * (dz * inv_r3) @ mass
    
    return np.stack([ax, ay, az], axis=1)

def simulate_step(pos, vel, mass, dt, G=1.0):
    """
    Velocity Verlet Integration Step
    """
    # 1. Half-step velocity
    acc = get_accelerations(pos, mass, G)
    vel_half = vel + acc * dt / 2.0
    
    # 2. Full-step position
    pos_new = pos + vel_half * dt
    
    # 3. Full-step velocity
    acc_new = get_accelerations(pos_new, mass, G)
    vel_new = vel_half + acc_new * dt / 2.0
    
    return pos_new, vel_new

# Example Initialization
N = 100
pos = np.random.randn(N, 3)
vel = np.random.randn(N, 3) * 0.1
mass = np.ones((N, 1))
dt = 0.01

# Run one step
pos, vel = simulate_step(pos, vel, mass, dt)
print(f"Updated positions (first 5):\n{pos[:5]}")
```
